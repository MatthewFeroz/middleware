## 2. Configure your API key

```python
import os, getpass

def _set_env(var: str):
    if not os.environ.get(var):
        os.environ[var] = getpass.getpass(f"{var}: ")

_set_env("OPENAI_API_KEY")
```

Here, we'll use LangSmith for tracing.

We'll log to a project, `middleware`.

```python
_set_env("LANGSMITH_API_KEY")
os.environ["LANGSMITH_TRACING"] = "true"
os.environ["LANGSMITH_PROJECT"] = "middleware"
```

## 3. Smoke test: build a tiny agent (no middleware yet)

If this runs and prints a response, you're ready for the lesson.

```python
from langchain.agents import create_agent

def get_weather(city: str) -> str:
    """Return the weather for a given city."""
    return f"The weather in {city} is sunny and 72°F."

agent = create_agent(
    model="openai:gpt-4o-mini",
    tools=[get_weather],
)

result = agent.invoke({"messages": [{"role": "user", "content": "What's the weather in Tokyo?"}]})
print(result["messages"][-1].content)
```
